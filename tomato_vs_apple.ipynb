```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import os
```

```python
path = "/content/drive/MyDrive/Colab Notebooks/archive/train/apples"
path2 = "/content/drive/MyDrive/Colab Notebooks/archive/train/tomatoes"
file_lst = os.listdir(path)
file_lst2 = os.listdir(path2)
print(file_lst)
print(file_lst2)
```

```python
file_dir=[]
for file in file_lst:
    filepath = path + '/' + file
    print(filepath)
    file_dir.append(filepath)

for file in file_lst2:
    filepath = path2 + '/' + file
    print(filepath)
    file_dir.append(filepath)
```

```python
print(type(file_dir))
print(len(file_dir))
file_dir[-215]
```

```python
img=[]
for i in file_dir:
  image =0
  image = cv2.imread(i, cv2.IMREAD_GRAYSCALE)
  image_100x100 = cv2.resize(image, (100,100)) # 이미지 크기를 100x100 픽셀로 변경
  img.append(image_100x100)
```

```python
imgg=np.array(img)
```

```python
imgg_2d = imgg.reshape(-1,100*100)
```

```python
imgg_2d.shape
```

```python
from sklearn.cluster import KMeans
km= KMeans(n_clusters=2, random_state=42)
km.fit(imgg_2d)
```

```python
print(km.labels_)
```

```python
print(np.unique(km.labels_,return_counts=True))
```

```python
def draw_img(arr, ratio=1):
  n=len(arr)

  rows = int(np.ceil(n/10))

  cols=n if rows < 2 else 10
  fig, axs = plt.subplots(rows,cols,figsize=(cols*ratio, rows*ratio), squeeze=False)
  for i in range(rows):
    for j in range(cols):
      if i*10 +j < n:
        axs[i,j].imshow(arr[i*10+j], cmap='gray')
      axs[i,j].axis('off')
  plt.show()
```

```python
draw_img(imgg[km.labels_==1])
```

```python
draw_img(imgg[km.labels_==0])
```
